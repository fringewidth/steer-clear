````python
import requests
import json
import csv
import time
import re
from typing import List, Dict
````

````python
import random
def call_gemini_api(prompt: str, api_key: str) -> str:
    """
    Call Gemini 2.0 Flash API via Google AI Studio with the given prompt
    """
    temp = max(0.3, min(1.0, random.gauss(0.7, 0.15)))
    
    # top_p: mean 0.9, variance ±0.05, clamp 0.8–0.95
    top_p = max(0.8, min(0.95, random.gauss(0.9, 0.03)))
    url = "https://generativelanguage.googleapis.com/v1beta/models/gemini-2.0-flash:generateContent"
    
    headers = {
        'Content-Type': 'application/json',
        'x-goog-api-key': api_key
    }
    
    data = {
        "contents": [
            {
                "parts": [
                    {
                        "text": prompt
                    }
                ]
            }
        ],
        "generationConfig": {
            "temperature": temp,
            "topP": top_p
        }
    }
    
    try:
        response = requests.post(url, headers=headers, json=data)
        response.raise_for_status()
        
        result = response.json()
        if 'candidates' in result and len(result['candidates']) > 0:
            return result['candidates'][0]['content']['parts'][0]['text']
        else:
            print(f"No candidates in response: {result}")
            return ""
            
    except requests.exceptions.RequestException as e:
        print(f"API request failed: {e}")
        return ""
    except (KeyError, IndexError) as e:
        print(f"Error parsing response: {e}")
        return ""
````

````python
def parse_prompts_from_response(response_text: str, category_name: str) -> List[Dict[str, str]]:
    """
    Parse the numbered prompts from Gemini's response and return as list of dicts
    """
    prompts = []
    
    # Split into lines and look for numbered items
    lines = response_text.split('\n')
    
    for line in lines:
        line = line.strip()
        # Match patterns like "1. prompt text" or "1) prompt text"
        match = re.match(r'^\d+[\.\)]\s*(.+)', line)
        if match:
            prompt_text = match.group(1).strip()
            # Remove any quotes around the prompt
            prompt_text = prompt_text.strip('"\'')
            if prompt_text:  # Only add non-empty prompts
                prompts.append({
                    'category': category_name,
                    'prompt': prompt_text
                })
    
    return prompts

def generate_category_prompts(category_instruction: str, category_name: str, api_key: str, target_count: int = 100) -> List[Dict[str, str]]:
    """
    Generate prompts for a specific category using batched API calls
    """
    all_prompts = []
    batch_size = 20  # Generate 20 prompts per API call
    
    print(f"Generating prompts for category: {category_name}")
    
    while len(all_prompts) < target_count:
        remaining = target_count - len(all_prompts)
        current_batch_size = min(batch_size, remaining)
        
        # Create the full prompt
        full_prompt = starting_prompt + category_instruction.replace("100", str(current_batch_size)) + output_format
        
        print(f"  Requesting {current_batch_size} prompts (total so far: {len(all_prompts)})")
        
        # Call the API
        response = call_gemini_api(full_prompt, api_key)
        
        if response:
            # Parse the response
            new_prompts = parse_prompts_from_response(response, category_name)
            all_prompts.extend(new_prompts)
            print(f"  Got {len(new_prompts)} prompts")
        else:
            print(f"  Failed to get response for batch")
            break
        
        # Add a small delay to be respectful to the API
        time.sleep(1)
        
        # Safety check to avoid infinite loop
        if len(all_prompts) >= target_count:
            break
    
    # Trim to exact target count if we got more
    return all_prompts[:target_count]
````

````python
def save_prompts_to_csv(all_prompts: List[Dict[str, str]], filename: str = "eval_temp.csv", append: bool = True):
    """
    Save all prompts to a CSV file (append mode by default)
    """
    import os
    
    file_exists = os.path.exists(filename)
    mode = 'a' if append and file_exists else 'w'
    
    print(f"File exists: {file_exists}, Mode: {mode}")
    print(f"{'Appending' if mode == 'a' else 'Writing'} {len(all_prompts)} prompts to {filename}")
    
    with open(filename, mode, newline='', encoding='utf-8') as csvfile:
        fieldnames = ['category', 'prompt']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        
        # Only write header if we're in write mode (new file or overwrite)
        if mode == 'w':
            writer.writeheader()
            
        for prompt_data in all_prompts:
            writer.writerow(prompt_data)
    
    print(f"Successfully {'appended to' if mode == 'a' else 'saved to'} {filename}")

# Extract category names from the category instructions
# category_names = [
#     "Moral/Ethical Reasoning",
#     "Emotional/Interpersonal Stance", 
#     "Analytical/Reasoning Style",
#     "Risk/Uncertainty Attitudes",
#     "Authority/Power Dynamics"
# ]


category_names = [
    "Historical Knowledge/Context",
    "Culinary/Baking Domain",
    "Controversial Social/Political Topics",
    "Technical/Scientific Questions",
    "Relationship/Social Advice"
]
````

````python
# Main execution script
def generate_evaluation_dataset(api_key: str, prompts_per_category: int = 20):
    """
    Generate the complete evaluation dataset
    """
    all_prompts = []
    
    print(f"Starting generation of evaluation dataset with {prompts_per_category} prompts per category")
    print("=" * 60)
    
    for i, (category_instruction, category_name) in enumerate(zip(categories, category_names)):
        print(f"\nProcessing category {i+1}/5: {category_name}")
        
        try:
            category_prompts = generate_category_prompts(
                category_instruction, 
                category_name, 
                api_key, 
                target_count=prompts_per_category
            )
            
            all_prompts.extend(category_prompts)
            print(f"✓ Successfully generated {len(category_prompts)} prompts for {category_name}")
            
        except Exception as e:
            print(f"✗ Error generating prompts for {category_name}: {e}")
            continue
    
    print("\n" + "=" * 60)
    print(f"Generation complete! Total prompts: {len(all_prompts)}")
    
    # Save to CSV (append mode)
    save_prompts_to_csv(all_prompts, append=True)
    
    return all_prompts
````
